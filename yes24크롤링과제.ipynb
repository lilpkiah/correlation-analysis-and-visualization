```python
from selenium import webdriver
import time
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlretrieve
def image_save(src,title):
    try:
        savename='C:/devtools/python_work/data/book_images/'+title+".jpg"
        urlretrieve(src,savename)
        print('저장됨')
    except:
        print("사진 오류")
```

```python
bookName=input("책 검색 : ")
url='http://www.yes24.com/searchcorner/Search?domain=BOOK&query='+quote(bookName)
driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(3) 
driver.get(url)
time.sleep(2)

i=0
n=4
book_list=[]

while True:
    html = driver.page_source
    soup = bs(html, 'html.parser')
    books=soup.select('div.goodsList.goodsList_list tr')
    
    for book in books:
        if i%2 ==0:
            title=book.select('td.goods_infogrp > p.goods_name.goods_icon > a >strong')[0].string
            print(title)
            
            #image=book.select('td.goods_img > a > img')
            #src=image[0]['src']
            #image_save(src,title)
            #print(src)
            
            book_info=book.select('div.goods_info')[0].text.split(' | ')
            author=book_info[0].strip()
            publisher=book_info[1].strip()
            year=book_info[2].strip()
            #print(author,publisher,year)
            price=book.select('div.goods_price > em')[0].text.replace(",","")
            #print(price)
            rating=book.select('span.gd_rating > em')
            if len(rating) == 0:
                rating='평점없음'
            else:
                rating=rating[0].text

            book_list.append(tuple([title,author,publisher,year,price,rating]))
        i+=1
    
    #print(len(books))
    #lastPage=driver.find_elements_by_css_selector('#schMid_wrap > div:nth-child(4) > div.goodsList.goodsList_list > div > a')
    #print(len(lastPage))
    
    if n==13:
        linkUrl=driver.find_element_by_css_selector('#schMid_wrap > div.goods_list_wrap.mgt30 > div.goodsList.goodsList_list > div > a.bw.ne')
        linkUrl.click()
        n=4
        time.sleep(2)
        continue
    
    linkUrl = driver.find_element_by_css_selector('#schMid_wrap > div.goods_list_wrap.mgt30 > div.goodsList.goodsList_list > div > a:nth-child('+str(n)+')')
    print(linkUrl.text)
    linkUrl.click()
    n+=1
    time.sleep(2)
        
    if len(books) < 40:
        break         

driver.quit()
print(book_list)
```
