## pairplot
- 그리그(grid) 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 표시
- 숫자형 column에 대해서만 그려준다

```python
sns.pairplot(tips)
plt.show()
```

```python
# hue 옵션으로 구분해서 볼 수도 있다.
sns.pairplot(tips, hue='size')
plt.show()
```

## lmplot
- lmplot은 column 간의 선형관계를 확인하기에 용이한 차트
- 또한, outlier도 같이 짐작해 볼 수 있다.

```python
sns.lmplot(x="total_bill", y="tip", height=8, data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', height=8, data=tips)
plt.show()
```

```python
# col_wrap으로 한 줄에 표기한 column의 갯수를 명시할 수 있다.
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='day', col_wrap=2, height=6, data=tips)
plt.show()
```

## jointplot
- scatter(산점도)와 histogram(분포)을 동시에 그려주며 숫자형 데이터만 표현 가능

```python
sns.jointplot(x="total_bill", y="tip", height=8, data=tips)
plt.show()
```

```python
# 선형관계를 표현하는 regression 라인 그리기도 가능.
# 옵션에 kind='reg'을 추가
sns.jointplot("total_bill", "tip", height=8, data=tips, kind="reg")
plt.show()
```
